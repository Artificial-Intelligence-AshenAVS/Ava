**Import Dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
#Loading data to pandas dataframes
hd_data = pd.read_csv('/content/heart_disease.csv')
```

```python
#1s 5 rows
hd_data.head()
```

```python
#data size of the dataset and info + checking missing values
hd_data.shape
hd_data.info()
hd_data.isnull().sum
```

```python
#statistical measures
hd_data.describe()
```

```python
#check distribution of target variable
hd_data['target'].value_counts()
```

**Split Features and Target**

```python
features = hd_data.drop(columns='target', axis=1)
tar = hd_data['target']

print(features)
print(tar)
```

**Splitting data -Training & Test**

```python
features_train, features_test, tar_train, tar_test = train_test_split(features, tar, test_size=0.2, stratify=tar, random_state=2)
```

```python
print(features.shape, features_train.shape, features_test.shape)
```

**Model tarining Process**

```python
model = LogisticRegression()
```

```python
model.fit(features_train, tar_train)
```

**Eveluation**

```python
#Training data
features_train_predict = model.predict(features_train)
training_acuracy = accuracy_score(features_train_predict, tar_train)
```

```python
print("Accuracy on training data:",training_acuracy)
```

```python
#Test data
features_test_predict = model.predict(features_test)
test_acuracy = accuracy_score(features_test_predict, tar_test)
```

```python
print("Accuracy on test data:",test_acuracy)
```
